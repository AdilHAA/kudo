```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
def get_denials_and_completed(sales_path: str, denials_path: str, groups_path: str):
    denials = pd.read_csv(denials_path)
    groups = pd.read_csv(groups_path)
    sales = pd.read_csv(sales_path)
    
    groups['Категория'] = groups['Категория'].fillna('Unk')
    groups = groups.rename(columns = {"Продукт": "SKU"})
    
    denials['Сумма_заказа'] = denials['Цена (р.)']*denials['Количество (шт.)']
    denials = denials.rename(columns = {"Артикул": "SKU"})
    denials = denials.fillna(0)
    result_denials = pd.merge(
    denials,
    groups,
    on='SKU',
    how='left'
    )
    result_denials = result_denials.fillna('UNK')
    
    sales['Сумма_заказа'] = sales['Цена (р.)']*(1-(sales['Скидка (%)']/100))*sales['Количество (шт.)']
    sales['Цена (р.)'] = sales['Цена (р.)']*(1-(sales['Скидка (%)']/100))
    sales = sales[sales['Скидка (%)'] >= 0]
    sales = sales.drop(columns=['Скидка (%)'])
    result_sales = pd.merge(
    sales,
    groups,
    on='SKU',
    how='left'
    )
    return result_sales,result_denials
    
```

```python
completed,denials = get_denials_and_completed('1-sales_train.csv','2-groups.csv','3-denials_train.csv')
```

```python
completed.to_csv('completed.csv')
denials.to_csv('denials.csv')
```
